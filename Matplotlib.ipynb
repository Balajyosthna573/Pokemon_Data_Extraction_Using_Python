```python
import numpy as np
import matplotlib.pyplot as plt
t = np.linspace(0,5,15)

plt.plot(t, t**2, color='coral', linestyle='--')
plt.plot(t, 4*t, color='indigo', linewidth=2)
```

```python
plt.subplot(1,3,1)
plt.plot(t,t,'c--')

plt.subplot(1,3,2)
plt.plot(t,t**2,'b--')

plt.subplot(1,3,3)
plt.plot(t,t**3,'mo')
```

```python
grades = {'John':70, 'ben':40, 'jyo':100, 'vicky':60, 'justin':50}
plt.bar(range(len(grades)), list(grades.values()), color='red')
plt.xticks(range(len(grades)), list(grades.keys()))
plt.xlabel('Students')
plt.xlabel('Grades')
```
